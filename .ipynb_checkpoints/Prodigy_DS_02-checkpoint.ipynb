```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load the data set #

data = pd.read_csv("/Users/macbook/Prodigy Info Tech Data Science Internship/titanic/test.csv")
# Load the dataset containing the 'Survived' column
survival_data = pd.read_csv('/Users/macbook/Prodigy Info Tech Data Science Internship/titanic/gender_submission.csv')
#View information on the dataset

print(data.columns)
print(survival_data.columns)

```

```python
# Merge the datasets on 'PassengerId'
data = pd.merge(data, survival_data, on='PassengerId')
print(data.head())
print(data.columns)
```

```python
print(data.isnull().sum())

```

```python
missing_cabin_percentage = data['Cabin'].isnull().mean() * 100
print(f"Percentage of missing Cabin values: {missing_cabin_percentage:.2f}%")
```

```python
# Fill missing age values with the median
data['Age'] = data['Age'].fillna(data['Age'].median())

# Drop rows with missing Embarked values
data = data.dropna(subset=['Cabin'])

# Fill missing Fare values with the median
data['Fare'] = data['Fare'].fillna(data['Fare'].median())
```

```python
print(data.isnull().sum())
```

```python
# Plot age distribution
sns.histplot(data['Age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()

# Count plot for survived
sns.countplot(x='Survived', data=data)
plt.title("Survival Count")
plt.show()

```

```python
sns.countplot(data=data, x='Survived', hue='Sex')
plt.title("Survival Count by Sex")
plt.show()
```

```python
sns.boxplot(data=data, x='Survived', y='Age')
plt.title("Age Distribution by Survival Status")
plt.show()
```

```python
sns.boxplot(data=data, x='Survived', y='Fare')
plt.title("Fare Distribution by Survival Status")
plt.show()

```

```python
sns.countplot(data=data, x='Survived', hue='Pclass')
plt.title("Survival Count by Passenger Class")
plt.show()
```

```python
# Group data by 'Sex' and calculate mean survival rate
survival_rate_by_sex = data.groupby('Sex')['Survived'].mean()


print(data[['Sex', 'Survived']].head())
print(data[['Sex', 'Survived']].info())

# Survival rate by class
sns.barplot(x='Pclass', y='Survived', data=data)
plt.title("Survival Rate by Class")
plt.show()
```

```python
# Select only numerical columns
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr_matrix = numerical_data.corr()

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot for Age, Fare, and Survived
sns.pairplot(data[['Age', 'Fare', 'Survived']])
plt.show()
```
